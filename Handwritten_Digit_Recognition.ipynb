```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Input
import matplotlib.pyplot as plt
```

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()

#Visualize an example
plt.imshow(X_train[0], cmap = 'gray')
plt.title(f"Label: {y_train[0]}")
plt.show()

#Normalise data
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
model = Sequential([
    Input(shape = (28,28)),
    Flatten(),
    Dense(128, activation = 'relu'),
    Dense(10, activation = 'softmax')
])
```

```python
# prepare model for training
model.compile(
    # adjusts weights to minimize errors
    optimizer = 'adam',
    #calculates error between predicted digit and actual digit
    loss = 'sparse_categorical_crossentropy',
    #outputs how accurate
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs = 5, validation_split = 0.2)
```

```python
#evaluates the test data using loss and accuracy
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print (f"Test Accuracy: {test_accuracy * 100: .2f}%")
```

```python
import numpy as np

#Number of test images to check
num_images = 5

for i in range(num_images):

  plt.imshow(X_test[i], cmap = 'gray')
  plt.title(f"Original Image {i+1}")
  plt.show()

  prediction = model.predict(X_test[i].reshape(1,28,28))
  print (f"Predicted Label: {np.argmax(prediction)}")
  print (f"Actual Label: {y_test[i]}")
  print ("-" * 30 )
```
